```python
import investpy
import datetime
import matplotlib.pyplot as plt
import pandas as pd

from mplfinance.original_flavor import candlestick_ohlc
import matplotlib.dates as mpl_dates
```

```python
search_result = investpy.search_quotes(text='apple', products=['stocks'],
                                       countries=['united states'])
t2 = datetime.datetime.now()
t1_stamp = (datetime.datetime.timestamp(t2)-86400*7)
t1 = datetime.datetime.fromtimestamp(t1_stamp)
df = search_result[0].retrieve_historical_data(
                                        from_date=t1.strftime('%d/%m/%Y'),
                                        to_date=t2.strftime('%d/%m/%Y'))
df['Date'] = df.index
print(df.head())
```

```python
df['date'] = df.index
# gca stands for 'get current axis'
ax = plt.gca()
df.plot(kind='line',x='date',y='Open', color='red', ax=ax)
#style1 = {'grid.linewidth': '0.8', 'grid.alpha':'0.5', 'grid.linestyle': 'solid', 'grid.color':     'b6b7b0'}
plt.style.use('classic')
#plt.style.use(style1)
#plt.grid()
plt.xlabel('Date')
plt.ylabel('Open cost')
plt.title('Paper cost')
plt.show()
```
